```python
# Graphing the number of different age ranges on each streaming platform
n = 4
r = np.arange(n)
width = 0.2
  
index = ['Netflix','Hulu','Prime Video','Disney+']

all_tot = [176, 124, 173, 129]
seven_tot = [299, 325, 213, 109]
thirteen_tot = [3,6,2,0]
sixteen_tot = [384, 501, 208, 21]
eighteen_tot = [482, 261, 164, 2]

barall = plt.bar(r, all_tot, width, color = 'r')
bar7 = plt.bar(r+width, seven_tot, width, color = 'b')
bar13 = plt.bar(r+width*2, thirteen_tot, width, color = 'c')
bar16 = plt.bar(r+width*2, sixteen_tot, width, color = 'm')
bar18 = plt.bar(r+width*3, eighteen_tot, width, color = 'y')

plt.xlabel("Streaming Service")
plt.ylabel("# Shows in Age Rating")
plt.title("Different Age Ratings for Various Streaming Services")
  

plt.xticks(r + width,index)
plt.legend(('all','7+','13+','16+','18+'))
  
plt.show()
```

```python
# Check which TV shows are on 2 or more streaming platfroms
index = ['Netflix', 'Hulu', 'Prime Video', 'Disney+']

row_sum = df_tv['Sum'] = df_tv[index].sum(axis = 1)

# Displays which platforms have the show
row1 = df_tv.loc[df_tv['Sum'] >= 2, index]
# Sum the instances that each platform has a show in common with another platform
row2 = df_tv.loc[df_tv['Sum'] >= 2, index].sum()
# Displays which shows are on 2 or more platforms
row3 = df_tv.loc[df_tv['Sum'] >= 2, 'Title']

print(row1)
print(row2)
print(row3)
```
