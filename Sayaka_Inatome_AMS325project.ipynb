```python
import pandas as pd
import numpy as np
```

```python
# organize dataframe
df = pd.read_csv('tv_shows.csv')
df = df.drop(columns={"Unnamed: 0"})
df.head()
```

```python
df.sort_values('Year', ascending=True)
# sort by oldest show
```

Does the date of release have an effect on whether or not the TV show is on a streaming service?


```python
# checking how many shows released in a certain year are on each streaming service
    # ex: 317 shows on Netflix were released in 2019
df1 = df.replace(0, np.nan, inplace=True) # replacing all the nan values to 0 as they have no effect here
df1 = df.groupby('Year').count()
df1 = df1.drop(columns={'ID', 'Title', 'Age', 'IMDb', 'Rotten Tomatoes', 'Type'})
df1
```

```python
Netflix = df1.drop(columns={'Hulu', 'Prime Video', 'Disney+'})
Netflix.sort_values('Netflix', ascending=False)
# sorting by most shows released in a year on Netflix
```

```python
Hulu = df1.drop(columns={'Netflix', 'Prime Video', 'Disney+'})
Hulu.sort_values('Hulu', ascending=False)
# sorting by most shows released in a year on Hulu
```

```python
PV = df1.drop(columns={'Hulu', 'Netflix', 'Disney+'})
PV.sort_values('Prime Video', ascending=False)
# sorting by most shows released in a year on Prime Video
```

```python
Disney = df1.drop(columns={'Hulu', 'Prime Video', 'Netflix'})
Disney.sort_values('Disney+', ascending=False)
# sorting by most shows released in a year on Disney+
```

Display information about a particular show

```python
name = input("Enter show name: ")

if (df[df["Title"].str.lower() == name.lower()]["ID"].count() != 0):
    print(df[df["Title"].str.lower() == name.lower()].to_string())
else:
    print(f"{name} not found")
```
